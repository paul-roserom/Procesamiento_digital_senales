# DETECCIÓN DE GESTOS EN SEÑALES EMG

Each gesture was recorded for 20 s with a sampling frequency of 1 KHz, allowing to obtain 20,000 samples per gesture. The sampling frequency was selected following the Nyquist sampling theorem, considering a maximum frequency for the EMG signal of 500 Hz.

[paper](https://www.mdpi.com/2076-3417/12/19/9700)
[dataset](https://zenodo.org/records/7637557)

* 0 = mano abierta
* 1 = puño
* 2 = ok
* 3 = indicando
* 4 = rock

```python
import pandas as pd # es una librería para manejo de diferentes tipos de archivos

# Si el separador es un espacio o tabulación (común en .txt)
# señal 1
señal_mano_abierta = "C:/Users/paul.rosero/Documents/UIDE/SEMANA_1/Clase/subject 1/0.txt"
# lectura del archivo en formato .txt
df = pd.read_csv(señal_mano_abierta, sep='\s+', names=['sensor1', 'sensor2', 'sensor3', 'sensor4'])
# imprime los primeros datos del archivo
print(df.head)
```

Graficar los datos de cada sensor

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(df['sensor1'], color = 'blue', label= 'Sensor 1')
plt.plot(df['sensor2'], color = 'black', label= 'Sensor 2')
plt.plot(df['sensor3'], color = 'red', label= 'Sensor 3')
plt.plot(df['sensor4'], color = 'green', label = 'Sensor 4')
plt.show
```

Generar sub-gráficas para cada sensor

```python
fig, axs = plt.subplots(4, 1, figsize=(10, 12), sharex=True)
axs[0].plot(df['sensor1'].iloc[0:1000], color = 'blue')
axs[0].set_title('Sensor 1')
axs[1].plot(df['sensor2'].iloc[0:1000], color = 'black')
axs[1].set_title('Sensor 2')
axs[2].plot(df['sensor3'].iloc[0:1000], color = 'red')
axs[2].set_title('Sensor 3')
axs[3].plot(df['sensor4'].iloc[0:1000], color = 'green')
axs[3].set_title('Sensor 4')
```

```python
import numpy as np 
# sumamos todos los sensores para tener una sola representación por movimiento 
mano_abierta = np.array(df['sensor1'])+np.array(df['sensor2'])+np.array(df['sensor3'])+np.array(df['sensor4'])
plt.plot(mano_abierta[0:1000])
```

```python
señal_puño = "C:/Users/paul.rosero/Documents/UIDE/SEMANA_1/Clase/subject 1/1.txt"
df = pd.read_csv(señal_puño, sep='\s+', names=['sensor1', 'sensor2', 'sensor3', 'sensor4'])
print(df.head)
```
